```python
import os
import random
```

```python
random.seed(42)
```

```python
egcm_dir = 'egcms/'
```

```python
pdb_ids = sorted([dir[:4] for dir in os.listdir(egcm_dir)])
```

```python
pdb_ids[:5]
```

```python
len(pdb_ids)
```

```python
random.shuffle(pdb_ids)
```

```python
pdb_ids[:5]
```

```python
train_set = pdb_ids[:10000]
validation_set = pdb_ids[10000:12000]
test_set = pdb_ids[12000:14000]
```

```python
with open('pdbbind_training_set.txt', 'w') as f :
    for pdb_id in train_set :
        f.write(pdb_id + '\n')
        
with open('pdbbind_val_set.txt', 'w') as f :
    for pdb_id in validation_set :
        f.write(pdb_id + '\n')
        
with open('pdbbind_test_set.txt', 'w') as f :
    for pdb_id in test_set :
        f.write(pdb_id + '\n')
```

```python
with open('pdbbind_training_set.txt') as f :
    pdb_ids = [pdb_id.strip() for pdb_id in f.readlines()]
```
